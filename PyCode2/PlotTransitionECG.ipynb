```python
def PlotTransitionECG(x, xph, mu_xph, var_xph, fs, target_file, file_name, target_noise_db, fntsiz, detail):
    x_f = np.concatenate(x, axis=0).tolist()  
    fig = plt.figure(figsize = (22, 12))
    plt.subplot(2, 2, 1)    
    #plt.title('ECG', fontsize = fntsiz)
    plt.xlabel("time [s]", fontsize = fntsiz)
    plt.xticks(fontsize = fntsiz)    
    plt.ylabel("amplitude [mV]", fontsize = fntsiz)    
    plt.yticks(fontsize = fntsiz)    
    plt.plot(np.arange(len(x_f[0:detail]))/fs, x_f[0:detail], lw = 4, label = 'time domain ECG')
    plt.legend(fontsize = fntsiz, loc = 1)
    current_values = plt.gca().get_yticks()
    ax = fig.gca()
    ax.set_yticklabels(['{:.1f}'.format(x) for x in current_values])
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    #-----
    plt.subplot(2, 2, 3)
    #plt.title('heartbeats', fontsize = fntsiz)
    plt.xlabel("time [s]", fontsize = fntsiz)
    plt.xticks(fontsize = fntsiz)
    plt.ylabel("amplitude [mV]", fontsize = fntsiz)  
    plt.yticks(fontsize = fntsiz)    
    plt.plot(np.arange(len(x[0]))/fs, x[0], lw = 4, label = 'time domain beats')
    plt.legend(handlelength=0, fontsize = fntsiz, loc = 1, frameon=False)
    for i in np.arange(1,len(x)):
        plt.plot(np.arange(len(x[i]))/fs, x[i], lw = 4)
    current_values = plt.gca().get_yticks()
    ax = fig.gca()
    ax.set_yticklabels(['{:.1f}'.format(x) for x in current_values])
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    #-----
    plt.subplot(2, 2, 4)
    #plt.title('heartbeats phase', fontsize = fntsiz)
    plt.xlabel("time [s]", fontsize = fntsiz)
    plt.xticks(fontsize = fntsiz)
    plt.ylabel("amplitude [mV]", fontsize = fntsiz)    
    plt.yticks(fontsize = fntsiz)
    plt.plot(np.arange(len(xph[0]))/fs, xph[0], lw = 4, label = 'phase domain beats')
    plt.legend(handlelength=0, fontsize = fntsiz, loc = 1, frameon = False)
    for i in np.arange(1,len(xph)):
        plt.plot(np.arange(len(xph[i]))/fs, xph[i], lw = 4)
    current_values = plt.gca().get_yticks()
    ax = fig.gca()
    ax.set_yticklabels(['{:.1f}'.format(x) for x in current_values])
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    #-----
    plt.subplot(2, 2, 2)
    #plt.title('average heartbeat phase', fontsize = fntsiz)
    plt.xlabel("time [s]", fontsize = fntsiz)
    plt.xticks(fontsize = fntsiz)
    plt.ylabel("amplitude [mV]", fontsize = fntsiz)    
    plt.yticks(fontsize = fntsiz)
    plt.plot(np.arange(len(mu_xph))/fs, mu_xph, lw = 4, color = 'magenta', label = 'phase domain \nmean beat')
    plt.legend(handlelength=2, fontsize = fntsiz, loc = 1)
    plt.fill_between(np.arange(len(mu_xph))/fs,
                     mu_xph - 3 * np.sqrt(var_xph),
                     mu_xph + 3 * np.sqrt(var_xph),
                     color = 'gray',
                     alpha = 0.2)
    current_values = plt.gca().get_yticks()
    ax = fig.gca()
    ax.set_yticklabels(['{:.1f}'.format(x) for x in current_values])
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    #-----
    fig.tight_layout()
    plt.savefig(target_file + file_name + '_' + str(target_noise_db) + 'dB.eps', format = 'eps', bbox_inches = 'tight')                
    plt.savefig(target_file + file_name + '_' + str(target_noise_db) + 'dB.png', format = 'png', bbox_inches = 'tight')        
    plt.show()
```
