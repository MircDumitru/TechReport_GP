```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,20,0.1)
s = np.sin(t)
target_snr_db = 10
var_n = np.mean(s**2) / (10**(target_snr_db/10))
n = np.random.normal(0, np.sqrt(var_n), len(s))    
x = s + n

fig = plt.figure(figsize = (22, 4))
#--------
plt.subplot(1,3,1)
plt.plot(s, label = 's')
plt.legend(loc = 1)
plt.subplot(1,3,2)
plt.plot(n, label = 'noise (var_n =' +  str('{:.3f}'.format(var_n)) + ')')
plt.legend(loc = 1)
plt.subplot(1,3,3)
plt.plot(x, label = 'x')
plt.legend(loc = 1)
#--------
fig.tight_layout()
# plt.savefig(target_file + file_name + '_' + str(target_noise_db) + 'dB.eps', format = 'eps', bbox_inches='tight')
# plt.savefig(target_file + file_name + '_'+ str(target_noise_db) + 'dB.png', format = 'png', bbox_inches='tight')
#--------
plt.show()
```
